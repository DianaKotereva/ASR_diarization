```
def preproc(text):
  # Убрать \n
  text = text.replace('\n', '').replace('Сотрудник 1', '').replace('Клиент 1', '').replace('Ж ', '').replace('С ', '').replace('Н ', '').replace('Ж[', '').replace('С[', '').replace('Н[', '')

  # Убрать таймстемпы
  # text = text[text.index(']')+1:]

  text = re.sub('[^А-ЯЁа-яё]', ' ', text)
  text = re.sub(' +', ' ', text)
  text = text.lower().strip()
  return text
```

```
preproced_texts_raw = [i for i in list(map(lambda x: preproc(x), raw_lines)) if len(i) > 0]
preproced_text_raw = ' '.join(preproced_texts_raw)

preproced_texts_etalon = [i for i in list(map(lambda x: preproc(x), etalon_lines)) if len(i) > 0]
preproced_text_etalon = ' '.join(preproced_texts_etalon)

print('WER', 100 * wer(preproced_text_etalon, preproced_text_raw))
print('CER', 100 * cer(preproced_text_etalon, preproced_text_raw))
```

```
clean_files = ['/content/egor_чистый.txt', '/content/my_sber_чистый.txt', '/content/sergey_sber_эталон.txt', '/content/tolstov_чистый.txt']
raw_files = ['/content/egor.txt', '/content/my_sber_сырой_диалог.txt', '/content/sergey_sber_сырой диалог.txt', '/content/tolstov.txt']

raw_text = ''
clean_text = ''

for clean_file, raw_file in zip(clean_files, raw_files):
  with open(raw_file, 'r') as file:
    raw_lines = file.readlines()

  with open(clean_file, 'r') as file:
    etalon_lines = file.readlines()

  preproced_texts_raw = [i for i in list(map(lambda x: preproc(x), raw_lines)) if len(i) > 0]
  preproced_text_raw = ' '.join(preproced_texts_raw)

  preproced_texts_etalon = [i for i in list(map(lambda x: preproc(x), etalon_lines)) if len(i) > 0]
  preproced_text_etalon = ' '.join(preproced_texts_etalon)

  raw_text += ' ' + preproced_text_raw
  clean_text += ' ' + preproced_text_etalon

raw_text = raw_text.strip()
clean_text = clean_text.strip()
```

```
print('WER', 100 * wer(clean_text, raw_text))
print('CER', 100 * cer(clean_text, raw_text))
```
